```python
import pandas as pd
import random 
import json

original_data_experiment_name = "fourth_experiment"
rephrased_data_experiment_name = "fifth_experiment"

number_of_house_sentences = 208
number_of_tech_sentences = 408

tech_file_path = f'data/experiments/{original_data_experiment_name}/{number_of_tech_sentences}_tech_sentences_with_prompts.csv'
house_file_path = f'data/experiments/{original_data_experiment_name}/{number_of_house_sentences}_house_sentences_with_prompts.csv'


tech_sentences_df = pd.read_csv(tech_file_path)
house_sentences_df = pd.read_csv(house_file_path)

tech_sentences_df
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
import ast

load_dotenv()

llm = ChatOpenAI(model="gpt-4-0613", temperature=0)

prompt = ChatPromptTemplate.from_messages([
    ("user", "{input}")
])

output_parser = StrOutputParser()

chain = prompt | llm | output_parser
```

```python
def construct_validation_prompt(df):
    prompts = [];
    
    for index, row in df.iterrows():
        prompt = ("Job-related knowledge is surrounded by ## symbols.  Job-related skill is surrounded by @@. "
                  "You have a sentence that is supposed to be from a job posting. \n"
                  f"Sentence: \"{row['results']}\"\n"
                  "You should examine if the text is coherent, makes sense, and could potentially be present in a real job posting.  Consider how well skills or knowledge fit the context. "
                  "If the text contains just skill or knowledge components without additional context, consider it coherent, Answer just \"Yes\" or \"No\".")
        
        prompts.append(prompt)
        
    return prompts
        
initial_tech_validation_prompts = construct_validation_prompt(tech_sentences_df)

initial_tech_validation_prompts
```

```python
def get_results(prompts, chain):
    results = []
    for current_prompt in prompts:
        result = chain.invoke({"input": current_prompt})
        results.append(result)
    return results

tech_initial_validation = get_results(initial_tech_validation_prompts, chain)
```

```python
initial_house_validation_prompts = construct_validation_prompt(house_sentences_df)

initial_house_validation_prompts
```

```python
house_initial_validation = get_results(initial_house_validation_prompts, chain)
```

```python
house_initial_validation.count("No")
```

```python
def construct_fix_coherence_prompts(df):
    prompts = [];
    
    for index, row in df.iterrows():
        prompt = (f"Job-related knowledge is surrounded by ## symbols. Job-related skill is surrounded by @@. "
                  f"You have a sentence that is supposed to be from a job posting. \n"
                  f"Sentence: \"{row['results']}\"\n"
                  "It lacks coherence. Rephrase it to make it sound less synthetic . "
                  "Ensure that all existing @@ and ## annotations remain unchanged. Don't create any new skills or knowledge components.")
        
        prompts.append(prompt)
        
    return prompts
```

```python
tech_df_is_coherent = pd.Series(tech_initial_validation)
tech_df_to_paraphrase = tech_sentences_df[tech_df_is_coherent == 'No'].copy()

fix_coherence_prompts = construct_fix_coherence_prompts(tech_df_to_paraphrase)
fix_coherence_prompts[0:20]
```

```python
fixed_tech_sentences = get_results(fix_coherence_prompts, chain)
```

```python
fixed_tech_sentences
```

```python
house_df_is_coherent = pd.Series(house_initial_validation)
house_df_to_paraphrase = house_sentences_df[house_df_is_coherent == 'No'].copy()

house_fix_coherence_prompts = construct_fix_coherence_prompts(house_df_to_paraphrase)
house_fix_coherence_prompts[0:10]
```

```python
fixed_house_sentences = get_results(house_fix_coherence_prompts, chain)
```

```python
def clean_results(res):
    cleaned_results = [s[1:-1] if s.startswith('"') and s.endswith('"') else s for s in res]
    cleaned_results = [s[1:-1] if s.startswith('\'') and s.endswith('\'') else s for s in cleaned_results]

    return cleaned_results

fixed_tech_sentences = clean_results(fixed_tech_sentences)

fixed_tech_sentences
```

```python
tech_df_to_paraphrase["results"] = fixed_tech_sentences
tech_df_to_paraphrase["rephrase_prompts"] = fix_coherence_prompts

final_tech_df = tech_sentences_df.copy()
final_tech_df["original_results"] = final_tech_df["results"]
final_tech_df.loc[tech_df_is_coherent == 'No', 'results'] = tech_df_to_paraphrase['results']
final_tech_df.loc[tech_df_is_coherent == 'No', 'rephrase_prompts'] = tech_df_to_paraphrase['rephrase_prompts']

final_tech_df
```

```python
final_tech_df.to_csv(f'data/experiments/{rephrased_data_experiment_name}/{number_of_tech_sentences}_tech_sentences_with_prompts.csv', index=False)
```

```python
house_df_to_paraphrase["results"] = fixed_house_sentences
house_df_to_paraphrase["rephrase_prompts"] = house_fix_coherence_prompts

final_house_df = house_sentences_df.copy()
final_house_df["original_results"] = final_house_df["results"]
final_house_df.loc[house_df_is_coherent == 'No', 'results'] = house_df_to_paraphrase['results']
final_house_df.loc[house_df_is_coherent == 'No', 'rephrase_prompts'] = house_df_to_paraphrase['rephrase_prompts']

final_house_df
```
