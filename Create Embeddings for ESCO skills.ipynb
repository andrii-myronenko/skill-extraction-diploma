```python
import redis

client = redis.Redis(host='localhost', port=6379, db=0)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('intfloat/e5-large-v2')

pipeline = client.pipeline()

for index, row in esco_dataset.iterrows():
    label = row['preferredLabel']
    description = row['description']
    embedding = model.encode(label).astype(np.float32).tolist()

    skill_data = {
        "label": label,
        "embedding": embedding,
        "description": description
    }

    redis_key = f"skill:{index:05}"
    pipeline.json().set(redis_key, "$", skill_data)

res = pipeline.execute()

print("Skills and embeddings stored in Redis.")
```

```python
from redis.commands.search.field import (
    NumericField,
    TextField,
    VectorField,
)
from redis.commands.search.indexDefinition import IndexDefinition, IndexType
from redis.commands.search.query import Query


schema = (
    TextField("$.label", no_stem=True, as_name="label"),
    VectorField(
        "$.embedding",
        "FLAT",
        {
            "TYPE": "FLOAT32",
            "DIM": vector_dimension,
            "DISTANCE_METRIC": "COSINE",
        },
        as_name="embedding",
    ),
    NumericField("$.index", as_name="index")
)

definition = IndexDefinition(prefix=["skill:"], index_type=IndexType.JSON)

res = client.ft("idx:skills").create_index(fields=schema, definition=definition)
```

```python
info = client.ft("idx:skills").info()

num_docs = info["num_docs"]
indexing_failures = info["hash_indexing_failures"]

print(f"{num_docs} documents indexed with {indexing_failures} failures")
```

```python
queries = ["work with gaming equipment"]
encoded_queries = model.encode(queries)
```

```python
def create_query_table(query, queries, encoded_queries, extra_params={}):
    results_list = []
    for i, encoded_query in enumerate(encoded_queries):
        result_docs = (
            client.ft("idx:skills")
            .search(
                query,
                {
                    "query_vector": np.array(
                        encoded_query, dtype=np.float32
                    ).tobytes()
                }
                | extra_params,
            )
            .docs
        )
        for doc in result_docs:
            vector_score = round(1 - float(doc.vector_score), 2)
            results_list.append(
                {
                    "query": queries[i],
                    "score": vector_score,
                    "label": doc.label,
                }
            )

    queries_table = pd.DataFrame(results_list)
    queries_table.sort_values(
        by=["query", "score"], ascending=[True, False], inplace=True
    )
    queries_table["query"] = queries_table.groupby("query")["query"].transform(
        lambda x: [x.iloc[0]] + [""] * (len(x) - 1)
    )
    queries_table.to_markdown(index=False)

    return queries_table
```

```python
query = (
    Query("(*)=>[KNN 3 @embedding $query_vector AS vector_score]")
    .sort_by("vector_score")
    .return_fields('vector_score', 'label')
    .dialect(2)
)


create_query_table(query, queries, encoded_queries)
```
