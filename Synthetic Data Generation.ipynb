```python
def get_random_occupation_uri(dataset):
    if not dataset.empty and 'conceptUri' in dataset.columns:
        random_index = random.randint(0, len(dataset) - 1)
        random_item = dataset.iloc[random_index]
        return random_item['conceptUri']
    else:
        return "Dataset is empty or does not contain the 'Occupation URI' column."

    
random_occupation_uri = get_random_occupation_uri(it_occupations)
random_occupation_uri
```

```python
import pandas as pd
import random

def get_random_skills_and_knowledge_uris(skill_occupation_dataset, occupation_uri, max_count=2):
    skills = skill_occupation_dataset[
        (skill_occupation_dataset["occupationUri"] == occupation_uri) &
        (skill_occupation_dataset["relationType"] == 'essential') &
        (skill_occupation_dataset["skillType"] == 'skill/competence')
    ]
    
    knowledge = skill_occupation_dataset[
        (skill_occupation_dataset["occupationUri"] == occupation_uri) &
        (skill_occupation_dataset["relationType"] == 'essential') &
        (skill_occupation_dataset["skillType"] == 'knowledge')
    ]
    
    total_selected = 0
    combined_uris = []

    num_skills_to_sample = random.randint(0, max_count)
    num_knowledge_to_sample = max_count - num_skills_to_sample
    
    if not skills.empty and num_skills_to_sample > 0:
        skills_sampled = skills.sample(n=min(num_skills_to_sample, len(skills)), replace=False)
        combined_uris.extend(skills_sampled['skillUri'].tolist())
        total_selected += len(skills_sampled)
    
    num_knowledge_to_sample = max_count - total_selected
    
    if not knowledge.empty and num_knowledge_to_sample > 0:
        knowledge_sampled = knowledge.sample(n=min(num_knowledge_to_sample, len(knowledge)), replace=False)
        combined_uris.extend(knowledge_sampled['skillUri'].tolist())
    
    return combined_uris

# Example usage
random_skills_knowledge = get_random_skills_and_knowledge_uris(
    esco_skills_occupations_dataset,
    random_occupation_uri
)

random_skills_knowledge
```

```python
def create_job_posting_prompt(occupation_uri, skill_uris, occupations_dataset, skills_dataset):
    occupation_name = occupations_dataset[occupations_dataset['conceptUri'] == occupation_uri]['preferredLabel'].iloc[0]
    
    matched_skills = skills_dataset[skills_dataset['conceptUri'].isin(skill_uris)]
    
    components = matched_skills.apply(lambda row: f"\n{row['skillType']} component: {row['preferredLabel']}".replace("skill/competence", "skill"), axis=1).tolist()
    components_str = ", ".join(components)
    
    prompt = (f"A knowledge component is understood as factual or theoretical information, including specifics, principles, and certifications relevant to a field. \n"
              "Skill component represents an ability or expertise, often demonstrated through verbs or adjective-noun combinations, focused on job-related tasks. \n"
              f"Given the abovementioned definitions write a sample sentence from a job posting for a \"{occupation_name}\" role. It should contain the following components: {components_str}. \n"
              "Annotate these components in your response, using '##' at the beginning and end for skills, and '@@' at the beginning and end for knowledge components. "
              "The response should not contain any additional skills and knowledge components, apart from the ones mentioned above. The response should be consise, yet creative.")

    return prompt

sample_prompt = create_job_posting_prompt(
    random_occupation_uri, 
    random_skills_knowledge, 
    esco_occupations_dataset, 
    esco_skills_dataset
)
print(sample_prompt)
```

```python
prompts = []
for _ in range(100):
    random_occupation_uri = get_random_occupation_uri(it_occupations)
    random_skill_uris = get_random_skills_and_knowledge_uris(
        esco_skills_occupations_dataset, 
        random_occupation_uri, 
    )
    prompt = create_job_posting_prompt(
        random_occupation_uri, 
        random_skill_uris, 
        esco_occupations_dataset, 
        esco_skills_dataset
    )
    prompts.append(prompt)
    
prompts
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-4-0613", temperature=0)

prompt = ChatPromptTemplate.from_messages([
    ("user", "{input}")
])

output_parser = StrOutputParser()
```

```python
results = []
for current_prompt in prompts:
    results.append(chain.invoke({"input": current_prompt}))
```

```python
def convert_to_bio_tagged_sentences(sentence):
    """
    Convert a sentence with special markers for skills (##) and knowledge (@@)
    into a list of tuples containing each word and its corresponding BIO tag.
    Each word is tagged twice, once for its role as a skill, and once for knowledge.
    """
    words = sentence.replace('##', ' ## ').replace('@@', ' @@ ').split()
    tagged_words = []
    
    skill_tag = 'O' 
    knowledge_tag = 'O' 

    for word in words:
        if word == '##':
            if skill_tag == 'O':
                skill_tag = 'B-Knowledge'  
            else:
                skill_tag = 'O'  
            continue  
        
        if word == '@@':
            if knowledge_tag == 'O':
                knowledge_tag = 'B-Skill'  
            else:
                knowledge_tag = 'O'  
            continue  
        
        tagged_words.append((word, skill_tag, knowledge_tag))

        if skill_tag == 'B-Skill':
            skill_tag = 'I-Skill'
        if knowledge_tag == 'B-Knowledge':
            knowledge_tag = 'I-Knowledge'
    
    return tagged_words
```

```python
bio_splitted_sentences = []

for cleaned_result in cleaned_results:
    bio_splitted_sentences.append(convert_to_bio_tagged_sentences(cleaned_result))
```

```python
def append_sentences_to_conll(filename, sentences_to_append):
    with open(filename, 'a') as file:
        for sentence in sentences_to_append:
            file.write("\n")
            for word, skill_tag, knowledge_tag in sentence:
                file.write(f"{word}\t{skill_tag}\t{knowledge_tag}\n")

# append_sentences_to_conll("data/SkillSpan_enriched/skillspan_tech_train.conll", bio_splitted_sentences)
```
