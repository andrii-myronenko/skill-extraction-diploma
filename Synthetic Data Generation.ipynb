```python
import pandas as pd

file_path = './data/ESCO/skills_en.csv'
esco_skills_dataset = pd.read_csv(file_path)

esco_skills_dataset
```

```python
file_path = './data/ESCO/occupations_en.csv'
esco_occupations_dataset = pd.read_csv(file_path)

esco_occupations_dataset
```

```python
file_path = './data/ESCO/occupationSkillRelations_en.csv'
esco_skills_occupations_dataset = pd.read_csv(file_path)

esco_skills_occupations_dataset
```

```python
ui_dev_row = esco_occupations_dataset[esco_occupations_dataset["preferredLabel"] == "user interface developer"]
ui_dev_row
```

```python
ui_dev_uri = ui_dev_row.iloc[0]["conceptUri"]

ui_dev_uri
```

```python
ui_dev_skills = esco_skills_occupations_dataset[esco_skills_occupations_dataset["occupationUri"] == ui_dev_uri]

ui_dev_skills
```

```python
ui_dev_skills
```

```python
merged_ui_def_skills_df = pd.merge(ui_dev_skills, esco_skills_dataset, left_on='skillUri', right_on='conceptUri')

merged_ui_def_skills_df = merged_ui_def_skills_df[merged_ui_def_skills_df["relationType"] == "essential"]

sampled_df = merged_ui_def_skills_df.sample(n=10)
```

```python
", ".join(sampled_df["preferredLabel"])
```

```python
from dotenv import load_dotenv
import ast

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4-0613", temperature=0)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are world class HR writer. Your goal is to write a job posting based on requried skills. Be creative and various in formats, if suitable insert required skill lists." +
     "Output just the posting. Invent the name of the company."),
    ("user", "{input}")
])
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
chain = prompt | llm | output_parser

chain.invoke({"input": "Write the job posting for the role of " + "user interface developer." + 
              "Include the following skills into the text: " + ", ".join(sampled_df["preferredLabel"]) + ". Make sure the skills are included without paraphrasing. Don't include any other skills"})
```
