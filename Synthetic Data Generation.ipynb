```python
def get_random_occupation_uri(dataset):
    """
    Selects a random item from the given dataset and returns its occupationUri.

    Parameters:
    - dataset (pd.DataFrame): The dataset containing occupation URIs.

    Returns:
    - str: A random occupation URI from the dataset.
    """

    if not dataset.empty and 'conceptUri' in dataset.columns:
        random_index = random.randint(0, len(dataset) - 1)
        random_item = dataset.iloc[random_index]
        return random_item['conceptUri']
    else:
        return "Dataset is empty or does not contain the 'Occupation URI' column."

    
random_occupation_uri = get_random_occupation_uri(it_occupations)
random_occupation_uri
```

```python
def get_random_skills_and_knowledge_uris(skill_occupation_dataset, occupation_uri, count_range=(1, 2)):
    """
    Retrieves 1 to 2 random skills and 1 to 2 knowledge components for a given occupation URI,
    ensuring at least one component is returned.

    Parameters:
    - skill_occupation_dataset (pd.DataFrame): The dataset containing occupation URIs and related skills/knowledge.
    - occupation_uri (str): The occupation URI for which to retrieve skills/knowledge.
    - count_range (tuple): The range (inclusive) of the number of skills/knowledge to retrieve. Default is (0, 2).

    Returns:
    - List[str]: A list containing the URIs of the randomly selected skills and knowledge components.
    """
    # Assuming a way to differentiate skills from knowledge, e.g., a column 'type' with values 'skill' or 'knowledge'
    skills = skill_occupation_dataset[
        (skill_occupation_dataset["occupationUri"] == occupation_uri) &
        (skill_occupation_dataset["relationType"] == 'essential') &
        (skill_occupation_dataset["skillType"] == 'skill/competence')
    ]
    
    knowledge = skill_occupation_dataset[
        (skill_occupation_dataset["occupationUri"] == occupation_uri) &
        (skill_occupation_dataset["relationType"] == 'essential') &
        (skill_occupation_dataset["skillType"] == 'knowledge')
    ]
    
    # Randomly sample skills and knowledge
    skills_sampled = skills.sample(n=random.randint(*count_range), replace=True) if not skills.empty else pd.DataFrame()
    knowledge_sampled = knowledge.sample(n=random.randint(*count_range), replace=True) if not knowledge.empty else pd.DataFrame()
    
    # Ensure at least one component is selected
    if skills_sampled.empty and knowledge_sampled.empty:
        if not skills.empty:
            skills_sampled = skills.sample(n=1)
        elif not knowledge.empty:
            knowledge_sampled = knowledge.sample(n=1)

    # Combine and return URIs
    combined_uris = pd.concat([skills_sampled, knowledge_sampled])['skillUri'].unique().tolist()
    return combined_uris


random_skills_knowledge = get_random_skills_and_knowledge_uris(
    esco_skills_occupations_dataset,
    random_occupation_uri
)

random_skills_knowledge
```

```python
def create_job_posting_prompt(occupation_uri, skill_uris, occupations_dataset, skills_dataset):
    """
    Creates a prompt for a job posting for a given occupation, including specified skills or knowledge components
    with appropriate annotations.

    Parameters:
    - occupation_uri (str): The URI of the occupation.
    - skill_uris (list): A list of skill URIs.
    - occupations_dataset (pd.DataFrame): The occupations dataset.
    - skills_dataset (pd.DataFrame): The skills dataset.

    Returns:
    - str: A crafted prompt for the job posting.
    """
    # Get the occupation name
    occupation_name = occupations_dataset[occupations_dataset['conceptUri'] == occupation_uri]['preferredLabel'].iloc[0]
    
    # Filter the skills dataset to find matches for the provided skill URIs
    matched_skills = skills_dataset[skills_dataset['conceptUri'].isin(skill_uris)]
    
    # Prepare labels and types for the matched skills, without pre-annotation
    components = matched_skills.apply(lambda row: f"{row['skillType']} component: {row['preferredLabel']}".replace("skill/competence", "skill"), axis=1).tolist()
    components_str = ", ".join(components)
    
    # Crafting the prompt with instructions for GPT to annotate
    prompt = (f"A knowledge component is understood as factual or theoretical information, including specifics, principles, and certifications pertinent to a field. "
              "Skill component represents an ability or expertise, often demonstrated through verbs or adjective-noun combinations, focused on job-related tasks. "
              f"Given the abovementioned definitions write a sample sentence from a job posting for a \"{occupation_name}\" role. It should contain the following components: {components_str}. "
              "Annotate these components in your response, using '##' at the beginning and end for skills, and '@@' at the beginning and end for knowledge components."
              "The response should not contain any additional skills and knowledge components, apart from the ones mentioned above. The response should not contain addional job details. The response should be consise, yet creative.")

    return prompt

sample_prompt = create_job_posting_prompt(
    random_occupation_uri, 
    random_skills_knowledge, 
    esco_occupations_dataset, 
    esco_skills_dataset
)
print(sample_prompt)
```

```python
prompts = []
for _ in range(100):
    random_occupation_uri = get_random_occupation_uri(it_occupations)
    random_skill_uris = get_random_skills_and_knowledge_uris(
        esco_skills_occupations_dataset, 
        random_occupation_uri, 
    )
    prompt = create_job_posting_prompt(
        random_occupation_uri, 
        random_skill_uris, 
        esco_occupations_dataset, 
        esco_skills_dataset
    )
    prompts.append(prompt)
    
prompts
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4-0613", temperature=0)
```

```python
def convert_to_bio_tagged_sentences(sentence):
    """
    Convert a sentence with special markers for skills (##) and knowledge (@@)
    into a list of tuples containing each word and its corresponding BIO tag.
    Each word is tagged twice, once for its role as a skill, and once for knowledge.
    """
    # Splitting the sentence into words, handling markers within words
    words = sentence.replace('##', ' ## ').replace('@@', ' @@ ').split()
    tagged_words = []
    
    skill_tag = 'O'  # Initialize skill tag state
    knowledge_tag = 'O'  # Initialize knowledge tag state

    for word in words:
        # Handle skill and knowledge markers
        if word == '##':
            if skill_tag == 'O':
                skill_tag = 'B-Skill'  # Begin skill entity
            else:
                skill_tag = 'O'  # End skill entity, reset tag
            continue  # Skip the marker itself
        
        if word == '@@':
            if knowledge_tag == 'O':
                knowledge_tag = 'B-Knowledge'  # Begin knowledge entity
            else:
                knowledge_tag = 'O'  # End knowledge entity, reset tag
            continue  # Skip the marker itself
        
        # Append the word with its current skill and knowledge tags
        tagged_words.append((word, skill_tag, knowledge_tag))

        # If currently within an entity, switch to the inside tag
        if skill_tag == 'B-Skill':
            skill_tag = 'I-Skill'
        if knowledge_tag == 'B-Knowledge':
            knowledge_tag = 'I-Knowledge'
    
    return tagged_words
```

```python
bio_splitted_sentences = []

for cleaned_validated_result in cleaned_validated_results:
    bio_splitted_sentences.append(convert_to_bio_tagged_sentences(cleaned_validated_result))
```

```python
def append_sentences_to_conll(filename, sentences_to_append):
    """
    Append given sentences to a CoNLL-format file, ensuring proper formatting.
    
    Args:
    - filename (str): Path to the CoNLL file.
    - sentences_to_append (list of list of tuples): Sentences to append, where each
      sentence is represented as a list of tuples, each containing a word followed
      by its skill tag and knowledge tag.
    """
    # Open the file to append
    with open(filename, 'a') as file:
        # For each sentence in the list
        for sentence in sentences_to_append:
            # Ensure each new sentence starts on a new line
            file.write("\n")  # This line ensures separation from previous content
            for word, skill_tag, knowledge_tag in sentence:
                # Write each word with its skill and knowledge tags, followed by a newline
                file.write(f"{word}\t{skill_tag}\t{knowledge_tag}\n")

append_sentences_to_conll("data/SkillSpan_enriched/skillspan_tech_train.conll", bio_splitted_sentences)
```
