```python
def get_random_occupation_uri(dataset):
    if not dataset.empty and 'conceptUri' in dataset.columns:
        random_index = random.randint(0, len(dataset) - 1)
        random_item = dataset.iloc[random_index]
        return random_item['conceptUri']
    else:
        return "Dataset is empty or does not contain the 'Occupation URI' column."

    
random_occupation_uri = get_random_occupation_uri(it_occupations)
random_occupation_uri
```

```python
def filter_json_data(data):
    house_data = []
    tech_data = []
    
    for item in data:
        if item['source'] == 'house':
            house_data.append(item)
        elif item['source'] == 'tech':
            tech_data.append(item)
    
    return house_data, tech_data

house_data, tech_data = filter_json_data(skillspan_raw_data)
```

```python
def annotate_sentence(tokens, tags_skill, tags_knowledge):
    final_tokens = []

    for index, (token, skill_tag, knowledge_tag) in enumerate(zip(tokens, tags_skill, tags_knowledge)):
        current_tokens = [];
        
        if skill_tag == 'B':
            current_tokens.append("@@");
        if knowledge_tag == 'B':
            current_tokens.append("##");
        current_tokens.append(token)
        
        if len(tokens) == index+1:
            if knowledge_tag == 'I' or knowledge_tag == 'B':
                current_tokens.append("##")
            if skill_tag == 'I' or skill_tag == 'B':
                current_tokens.append("@@")
        else:
            next_knowledge_tag = tags_knowledge[index+1]
            if (knowledge_tag == 'I' or knowledge_tag == 'B') and (next_knowledge_tag == 'O' or next_knowledge_tag == 'B'):
                current_tokens.append("##")
            next_skill_tag = tags_skill[index+1]
            if (skill_tag == 'I' or skill_tag == 'B') and (next_skill_tag == 'O' or next_skill_tag == 'B'):
                current_tokens.append("@@")
        
        final_tokens.append(''.join(current_tokens))

    return ' '.join(final_tokens)

     
idx = 105;

print(tech_data[idx])
annotate_sentence(tech_data[idx]['tokens'], tech_data[idx]['tags_skill'], tech_data[idx]['tags_knowledge'])
```

```python
def get_random_skill_or_knowledge_urls(skill_occupation_dataset, occupation_uri, skill_type, n=1):
    essential_skills = skill_occupation_dataset[
        (skill_occupation_dataset["occupationUri"] == occupation_uri) &
        (skill_occupation_dataset["relationType"] == 'essential') &
        (skill_occupation_dataset["skillType"] == skill_type)
    ]
    
    num_essential_skills = len(essential_skills)
    
    if num_essential_skills >= n:
        knowledge_sampled = essential_skills.sample(n, replace=False)
    else:
        optional_skills = skill_occupation_dataset[
            (skill_occupation_dataset["occupationUri"] == occupation_uri) &
            (skill_occupation_dataset["relationType"] == 'optional') &
            (skill_occupation_dataset["skillType"] == skill_type)
        ]

        num_optional_skills_needed = n - num_essential_skills
        optional_skills_sampled = optional_skills.sample(num_optional_skills_needed, replace=True)

        knowledge_sampled = pd.concat([essential_skills, optional_skills_sampled])

    return knowledge_sampled['skillUri']

def get_random_knowledge_urls(skill_occupation_dataset, occupation_uri, n = 1):
    return get_random_skill_or_knowledge_urls(
        skill_occupation_dataset,
        occupation_uri,
        'knowledge',
        n
    )


def get_random_skill_urls(skill_occupation_dataset, occupation_uri, n=1):
    return get_random_skill_or_knowledge_urls(
        skill_occupation_dataset,
        occupation_uri,
        'skill/competence',
        n
    )


random_skills = get_random_skill_urls(
    esco_skills_occupations_dataset,
    random_occupation_uri,
    3
)

random_skills
```

```python
def build_annotation_prompt(n_skills, n_knowledge):
    prompt = ""
    if (n_skills > 0):
        prompt = prompt + "part surrounded by @@ is a job-related skill"
        if (n_knowledge > 0):
            prompt = prompt + ", and a "
    if (n_knowledge > 0):
            prompt = prompt + "part surrounded by ## is job-related knowledge"
            
    return prompt

def remove_ict_it(string):
    # Remove "ICT" and "IT" from the string (case-insensitive)
    modified_string = string.replace("ICT", "").replace("IT", "")
    
    # Remove any extra whitespace that may have been introduced
    modified_string = " ".join(modified_string.split())
    
    return modified_string

def build_entity_description_prompt(n_skills, n_knowledge):
    prompt = ""
    if (n_skills > 0):
        prompt = prompt + "the skills"
        if (n_knowledge > 0):
            prompt = prompt + " and "
    if (n_knowledge > 0):
            prompt = prompt + "the knowledge"
            
    return prompt

def build_new_entities_prompt(n_skills, n_knowledge, skills_dataset, skill_occupation_dataset, occupation_uri):
    prompt = ""
    
    if(n_skills > 0):
        random_skills = get_random_skill_urls(skill_occupation_dataset, occupation_uri, n_skills)
        random_skills = esco_skills_dataset[esco_skills_dataset['conceptUri'].isin(random_skills)]['preferredLabel']
        prompt = prompt + "for skills: " + ", ".join([f"@@{remove_ict_it(skill)}@@" for skill in random_skills])

        if (n_knowledge > 0):
            prompt = prompt + ", and "
    
    if (n_knowledge > 0):
        random_knowledge = get_random_knowledge_urls(skill_occupation_dataset, occupation_uri, n_knowledge)
        random_knowledge = esco_skills_dataset[esco_skills_dataset['conceptUri'].isin(random_knowledge)]['preferredLabel']
        prompt = prompt + "for knowledge: " + ", ".join([f"##{remove_ict_it(knowledge)}##" for knowledge in random_knowledge])
            
    return prompt
```

```python
def create_job_posting_prompt(original_sentence, n_skills, n_knowledge, occupation_uri, skills_dataset, skill_occupation_dataset, occupations_dataset):
    occupation_name = occupations_dataset[occupations_dataset['conceptUri'] == occupation_uri]['preferredLabel'].iloc[0]
    
    # Crafting the prompt with instructions for GPT to annotate
    prompt = (f"Take this sentence from a Job Posting.\n\n"
              f"\"{original_sentence}\"\n\n"
              f"A {build_annotation_prompt(n_skills, n_knowledge)}. "
              f"Write a sentence preserving annotations, but instead of {build_entity_description_prompt(n_skills, n_knowledge)}, you should put the following entities: "
              f"{build_new_entities_prompt(n_skills, n_knowledge, skills_dataset, skill_occupation_dataset, occupation_uri)}. "
              f"You may rephrase the sentence a little to accomodate new entities. Make sure not to add any other additional job details in the generated sentence, apart from the ones mentioned above.")

    return prompt
```

```python
def generate_prompts(
    reference_data, 
    source_occupations, 
    skills_dataset, 
    skills_occupations_dataset, 
    occupations_dataset, 
    num_prompts
):
    prompts = []
    
    for _ in range(num_prompts):
        random_occupation_uri = get_random_occupation_uri(source_occupations)
        random_reference = random.choice(reference_data)
        
        prompt = create_job_posting_prompt(
            random_reference["highlighted_sentence"],
            random_reference["count_skills"],
            random_reference["count_knowledge"],
            random_occupation_uri,
            skills_dataset,
            skills_occupations_dataset,
            occupations_dataset
        )
        
        prompts.append(prompt)
    
    return prompts

# Example usage
num_prompts = 300
prompts = generate_prompts(
    tech_data_with_entities,
    it_occupations,
    esco_skills_dataset,
    esco_skills_occupations_dataset,
    esco_occupations_dataset,
    num_prompts
)

prompts[0:5]
```

```python
def convert_to_bio_tagged_sentences(sentence):
    """
    Convert a sentence with special markers for skills (##) and knowledge (@@)
    into a list of tuples containing each word and its corresponding BIO tag.
    Each word is tagged twice, once for its role as a skill, and once for knowledge.
    """
    # Splitting the sentence into words, handling markers within words
    words = sentence.replace('##', ' ## ').replace('@@', ' @@ ').split()
    tagged_words = []
    
    skill_tag = 'O'  # Initialize skill tag state
    knowledge_tag = 'O'  # Initialize knowledge tag state

    for word in words:
        # Handle skill and knowledge markers
        if word == '##':
            if skill_tag == 'O':
                skill_tag = 'B-Skill'  # Begin skill entity
            else:
                skill_tag = 'O'  # End skill entity, reset tag
            continue  # Skip the marker itself
        
        if word == '@@':
            if knowledge_tag == 'O':
                knowledge_tag = 'B-Knowledge'  # Begin knowledge entity
            else:
                knowledge_tag = 'O'  # End knowledge entity, reset tag
            continue  # Skip the marker itself
        
        # Append the word with its current skill and knowledge tags
        tagged_words.append((word, skill_tag, knowledge_tag))

        # If currently within an entity, switch to the inside tag
        if skill_tag == 'B-Skill':
            skill_tag = 'I-Skill'
        if knowledge_tag == 'B-Knowledge':
            knowledge_tag = 'I-Knowledge'
    
    return tagged_words
```

```python
bio_splitted_sentences = []

for cleaned_result in cleaned_results:
    bio_splitted_sentences.append(convert_to_bio_tagged_sentences(cleaned_result))
    
bio_splitted_sentences[:5]
```

```python
def append_sentences_to_conll(filename, sentences_to_append):
    """
    Append given sentences to a CoNLL-format file, ensuring proper formatting.
    
    Args:
    - filename (str): Path to the CoNLL file.
    - sentences_to_append (list of list of tuples): Sentences to append, where each
      sentence is represented as a list of tuples, each containing a word followed
      by its skill tag and knowledge tag.
    """
    # Open the file to append
    with open(filename, 'a') as file:
        # For each sentence in the list
        for sentence in sentences_to_append:
            # Ensure each new sentence starts on a new line
            file.write("\n")  # This line ensures separation from previous content
            for word, skill_tag, knowledge_tag in sentence:
                # Write each word with its skill and knowledge tags, followed by a newline
                file.write(f"{word}\t{skill_tag}\t{knowledge_tag}\n")

append_sentences_to_conll("data/SkillSpan_enriched/skillspan_tech_train.conll", bio_splitted_sentences)
```
