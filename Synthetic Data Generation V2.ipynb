```python
def get_random_occupation_uri(occupations_df, relations_df, min_skills, min_knowledge):
    skills_knowledge_counts = relations_df.groupby(['occupationUri', 'skillType']).size().unstack(fill_value=0)
    
    # Find occupations that meet the minimum skill and knowledge requirements
    qualified_occupations = skills_knowledge_counts[
        (skills_knowledge_counts.get('skill/competence', 0) >= min_skills) & 
        (skills_knowledge_counts.get('knowledge', 0) >= min_knowledge)
    ].index.tolist()
    
    if not qualified_occupations:
        raise ValueError("No occupations meet the specified criteria of at least {} skills and {} knowledge components.".format(min_skills, min_knowledge))
    
    # Filter the original occupations DataFrame to include only qualified occupations
    qualified_occupations_df = occupations_df[occupations_df['conceptUri'].isin(qualified_occupations)]
    
    if qualified_occupations_df.empty:
        raise ValueError("No occupations in the occupations dataframe meet the criteria.")

    # Randomly select an occupation URI from the filtered dataframe
    selected_occupation_uri = random.choice(qualified_occupations_df['conceptUri'].tolist())
    
    return selected_occupation_uri

random_occupation_uri = get_random_occupation_uri(it_occupations, esco_skills_occupations_dataset, 15, 15)
random_occupation_uri
```

```python
def filter_json_data(data):
    house_data = []
    tech_data = []
    
    for item in data:
        if item['source'] == 'house':
            house_data.append(item)
        elif item['source'] == 'tech':
            tech_data.append(item)
    
    return house_data, tech_data

house_data, tech_data = filter_json_data(skillspan_raw_data)
```

```python
def annotate_sentence(tokens, tags_skill, tags_knowledge):
    final_tokens = []

    for index, (token, skill_tag, knowledge_tag) in enumerate(zip(tokens, tags_skill, tags_knowledge)):
        current_tokens = [];
        
        if skill_tag == 'B':
            current_tokens.append("@@");
        if knowledge_tag == 'B':
            current_tokens.append("##");
        current_tokens.append(token)
        
        if len(tokens) == index+1:
            if knowledge_tag == 'I' or knowledge_tag == 'B':
                current_tokens.append("##")
            if skill_tag == 'I' or skill_tag == 'B':
                current_tokens.append("@@")
        else:
            next_knowledge_tag = tags_knowledge[index+1]
            if (knowledge_tag == 'I' or knowledge_tag == 'B') and (next_knowledge_tag == 'O' or next_knowledge_tag == 'B'):
                current_tokens.append("##")
            next_skill_tag = tags_skill[index+1]
            if (skill_tag == 'I' or skill_tag == 'B') and (next_skill_tag == 'O' or next_skill_tag == 'B'):
                current_tokens.append("@@")
        
        final_tokens.append(''.join(current_tokens))

    return ' '.join(final_tokens)

     
idx = 105;

print(tech_data[idx])
annotate_sentence(tech_data[idx]['tokens'], tech_data[idx]['tags_skill'], tech_data[idx]['tags_knowledge'])
```

```python
def annotate_data(data):
    for idx in range(len(data)):
        data[idx]['highlighted_sentence'] = annotate_sentence(data[idx]['tokens'], data[idx]['tags_skill'], data[idx]['tags_knowledge'])
        data[idx]['count_skills'] = data[idx]['tags_skill'].count("B")
        data[idx]['count_knowledge'] = data[idx]['tags_knowledge'].count("B")


annotate_data(tech_data)
annotate_data(house_data)
```

```python
def get_random_skill_or_knowledge_urls(skill_occupation_dataset, occupation_uri, skill_type, n=1):
    skills = skill_occupation_dataset[
        (skill_occupation_dataset["occupationUri"] == occupation_uri) &
        (skill_occupation_dataset["skillType"] == skill_type)
    ]
    
    num_skills = len(skills)
    
    if num_skills >= n:
        skills_sampled = skills.sample(n, replace=False)
    else:
        skills_sampled = skills.sample(num_skills, replace=True)

    return skills_sampled['skillUri']

def get_random_knowledge_urls(skill_occupation_dataset, occupation_uri, n = 1):
    return get_random_skill_or_knowledge_urls(
        skill_occupation_dataset,
        occupation_uri,
        'knowledge',
        n
    )


def get_random_skill_urls(skill_occupation_dataset, occupation_uri, n=1):
    return get_random_skill_or_knowledge_urls(
        skill_occupation_dataset,
        occupation_uri,
        'skill/competence',
        n
    )


random_skills = get_random_skill_urls(
    esco_skills_occupations_dataset,
    random_occupation_uri,
    3
)

random_skills
```

```python
def build_annotation_prompt(n_skills, n_knowledge):
    prompt = ""
    if (n_skills > 0):
        prompt = prompt + "part surrounded by @@ is a job-related skill"
        if (n_knowledge > 0):
            prompt = prompt + ", and a "
    if (n_knowledge > 0):
            prompt = prompt + "part surrounded by ## is job-related knowledge"
            
    return prompt

def build_entity_description_prompt(n_skills, n_knowledge):
    prompt = ""
    if (n_skills > 0):
        prompt = prompt + "the skills"
        if (n_knowledge > 0):
            prompt = prompt + " and "
    if (n_knowledge > 0):
            prompt = prompt + "the knowledge"
            
    return prompt

def build_new_entities_prompt(n_skills, n_knowledge, skills_dataset, skill_occupation_dataset, occupation_uri):
    prompt = ""
    used_skills = []
    used_knowledge = []
    
    if(n_skills > 0):
        random_skills = get_random_skill_urls(skill_occupation_dataset, occupation_uri, n_skills)
        random_skills = esco_skills_dataset[esco_skills_dataset['conceptUri'].isin(random_skills)]['preferredLabel'].tolist()
        prompt = prompt + "for skills: " + ", ".join([f"@@{skill}@@" for skill in random_skills])
        used_skills = random_skills  
        
        if (n_knowledge > 0):
            prompt = prompt + ", and "
    
    if (n_knowledge > 0):
        random_knowledge = get_random_knowledge_urls(skill_occupation_dataset, occupation_uri, n_knowledge)
        random_knowledge = esco_skills_dataset[esco_skills_dataset['conceptUri'].isin(random_knowledge)]['preferredLabel'].tolist()
        prompt = prompt + "for knowledge: " + ", ".join([f"##{knowledge}##" for knowledge in random_knowledge])
        used_knowledge = random_knowledge
            
    return prompt, used_skills, used_knowledge
```

```python
def create_job_posting_prompt(original_sentence, n_skills, n_knowledge, occupation_uri, skills_dataset, skill_occupation_dataset, occupations_dataset):
    occupation_name = occupations_dataset[occupations_dataset['conceptUri'] == occupation_uri]['preferredLabel'].iloc[0]
    new_enities_prompt, used_skills, used_knowledge,  = build_new_entities_prompt(n_skills, n_knowledge, skills_dataset, skill_occupation_dataset, occupation_uri)
    
    prompt = (f"Take this sentence from a Job Posting.\n\n"
              f"\"{original_sentence}\"\n\n"
              f"A {build_annotation_prompt(n_skills, n_knowledge)}. "
              f"Write a sentence preserving annotations, but instead of {build_entity_description_prompt(n_skills, n_knowledge)}, you should put the following entities: "
              f"{new_enities_prompt}."
              f" Only adjust grammar minimally to maintain the readability and coherence of the sentence with the new entities. ")

    return prompt, used_skills, used_knowledge, occupation_name
```

```python
def generate_prompts(
    reference_data, 
    source_occupations, 
    skills_dataset, 
    skills_occupations_dataset, 
    occupations_dataset, 
    num_prompts
):
    prompts_data = []

    references = random.sample(reference_data, num_prompts)
    
    for random_reference in references:
        random_occupation_uri = get_random_occupation_uri(
            source_occupations, 
            skills_occupations_dataset,
            random_reference["count_skills"],
            random_reference["count_knowledge"]
        )

        prompt, used_skills, used_knowledge, occupation_name  = create_job_posting_prompt(
            random_reference["highlighted_sentence"],
            random_reference["count_skills"],
            random_reference["count_knowledge"],
            random_occupation_uri,
            skills_dataset,
            skills_occupations_dataset,
            occupations_dataset
        )
        
        if (len(used_skills) == 0 and len(used_knowledge) == 0):
            print(prompt)
        
        prompts_data.append({
            "occupation_name": occupation_name,
            "skills": used_skills,
            "knowledge": used_knowledge,
            "prompt": prompt
        })

    
    return pd.DataFrame(prompts_data)


num_prompts = 408

prompts_df = generate_prompts(
    tech_data_with_entities,
    it_occupations,
    esco_skills_dataset,
    esco_skills_occupations_dataset,
    esco_occupations_dataset,
    num_prompts
)

prompts_df
```

```python
def get_results(prompts, chain):
    results = []
    for current_prompt in prompts:
        result = chain.invoke({"input": current_prompt})
        results.append(result)
    return results

results = get_results(prompts_df['prompt'].tolist(), chain)
```

```python
def clean_results(res):
    cleaned_results = [s[1:-1] if s.startswith('"') and s.endswith('"') else s for s in res]
    cleaned_results = [s[1:-1] if s.startswith('\'') and s.endswith('\'') else s for s in cleaned_results]

    return cleaned_results

cleaned_results = clean_results(results)

cleaned_results[:5]
```

```python
def generate_rewrite_prompts(data, n):
    # Check if the requested number of sentences is more than available
    n = min(n, len(data))
    
    # Randomly select n entries from the data
    selected_entries = random.sample(data, n)
    
    prompts = []
    for entry in selected_entries:
        sentence = entry['highlighted_sentence']
        prompt = (f"Take this sentence from a Job Posting.\n\n"
          f"Sentence: \"{sentence}\"\n\n"
          f"Rewrite it by finding synonyms but not changing the meaning. "
          f"Don't change the style of the text. The sentence should look like it is from the vacancy.")
        prompts.append(prompt)
    
    return prompts
```
